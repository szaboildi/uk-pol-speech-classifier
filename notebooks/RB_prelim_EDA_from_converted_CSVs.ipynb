```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## UK data

```python
uk_data = pd.read_feather("/content/drive/MyDrive/Le Wagon Final Project /Corp_HouseOfCommons_V2.feather")
```

```python
uk_data["terms"] = uk_data["terms"].astype(int)
uk_data = uk_data[uk_data["terms"] >= 50]
uk_data["word_n_full"] = uk_data.apply(lambda row: len(row["text"].strip().split()), axis=1)
uk_data["date"] = pd.to_datetime(uk_data["date"]).dt.year
```

```python
uk_data["word_n_full"] = uk_data.apply(lambda row: len(row["text"].strip().split()), axis=1)
```

```python
uk_data_150 = uk_data[uk_data["word_n_full"] >= 150]
```

```python
uk_data_150.groupby("party").size().reset_index(name="n_speeches").\
sort_values("n_speeches", ascending=False).reset_index(drop=True)
```

```python
sns.histplot(uk_data_150["word_n_full"])
plt.xlim(0, 4000)
plt.xlabel('Number of words');
```

```python
sns.histplot(uk_data_150["date"]);
```

```python
datasets = [uk_data_150, uk_data_200, uk_data_250, uk_data_300,
            uk_data_350, uk_data_400, uk_data_450, uk_data_500]

# Define colors for each histogram
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray']

# Plot each histogram with a loop
for i, data in enumerate(datasets):
    sns.histplot(data["word_n_full"], color=colors[i], label=f'{i*50+150}')

# Set x-axis limit and label
plt.xlim(0, 4000)
plt.xlabel('Number of words')

# Display legend
plt.legend()

# Display the plot
plt.show()
```
