```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = pd.read_csv("../processed_data/features_1000sample_400min_600cutoff.csv", index_col="Unnamed: 0")
```

```python
y_df = pd.read_csv("../processed_data/target_1000sample_400min_600cutoff.csv", index_col="Unnamed: 0")
```

```python
codes = {p: i for i, p in enumerate(y_df["party"].unique())}
print(codes)

y = y_df["party"].map(codes)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```
